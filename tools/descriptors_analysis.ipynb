```python
import os

path = "/home/mjustyna/RNA-GNN/samples/af3_preds/pdbs"
files = os.listdir(path)
files = [f for f in files if f.endswith(".pdb")]
files = [f.replace('_AA.pdb', '') for f in files]
files = [f.replace('.pdb', '') for f in files]

# print(files)
# print("--------")
with open("all-output.txt", 'r') as f:
    lines = f.readlines()

lines = [l.split() for l in lines]
lines = [l for l in lines if len(l)>=7 and len(l[1])>=12]
pairs = dict([(l[1], int(l[2])) for l in lines])
seqs = dict([(l[1], l[-pairs[l[1]]:]) for l in lines])
len_seqs = dict([(k, len("".join(v))) for k, v in seqs.items()])
pairs
```

```python
ones = [f for f in files if pairs[f] == 1]
twos = [f for f in files if pairs[f] == 2]
threes = [f for f in files if pairs[f] == 3]
fours = [f for f in files if pairs[f] == 4]
fives = [f for f in files if pairs[f] == 5]
print(f'Ones: {ones}')
print(f'Twos: {twos}')
print(f'Threes: {threes}')
print(f'Fours: {fours}')
print(f'Fives: {fives}')

one_lens = [len_seqs[f] for f in ones]
two_lens = [len_seqs[f] for f in twos]
three_lens = [len_seqs[f] for f in threes]
four_lens = [len_seqs[f] for f in fours]
five_lens = [len_seqs[f] for f in fives]

# print(f'One lens: {min(one_lens)}, {max(one_lens)}')
# print(f'Two lens: {min(two_lens)}, {max(two_lens)}')
# print(f'Three lens: {min(three_lens)}, {max(three_lens)}')
# print(f'Four lens: {min(four_lens)}, {max(four_lens)}')
print(f'Five lens: {min(five_lens)}, {max(five_lens)}')
```

```python
all_ones = [f for f in pairs.keys() if pairs[f] == 1]
all_threes = [f for f in pairs.keys() if pairs[f] == 3]
all_fours = [f for f in pairs.keys() if pairs[f] == 4]
all_fives = [f for f in pairs.keys() if pairs[f] == 5]
print(len(all_ones))
my_preds_path = "/home/mjustyna/RNA-GNN/samples/glowing-terrain-25-eval/800"
my_preds = os.listdir(my_preds_path)
my_preds = [f for f in my_preds if f.endswith("_AA.pdb")]
my_preds = [f.replace('-000001_AA.pdb', '') for f in my_preds]

my_ones = set([f for f in my_preds if f in all_ones])
my_threes = set([f for f in my_preds if f in all_threes])
print(len(my_ones))
ones = set(ones)
my_ones = my_ones.difference(ones)
print(len(my_ones))

threes = set(threes)
my_threes = my_threes.difference(threes)
print(len(my_threes))

# sample random 16 sequences from my_ones
import random
# set seed
random.seed(42)
my_ones = list(my_ones)
random.shuffle(my_ones)
my_ones = my_ones[:16]
my_ones.extend(ones)

my_threes = list(my_threes)
random.shuffle(my_threes)
my_threes = my_threes[:14]
my_threes.extend(threes)
print(len(my_ones), len(my_threes))
print(my_threes)

fours = random.shuffle(all_fours)
fours = all_fours[:20]

fives = random.shuffle(all_fives)
fives = all_fives[:20]
print(len(fours), len(fives))
```
