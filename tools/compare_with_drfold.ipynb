```python
import os

rna3db_train_dir = "/home/mjustyna/RNA-GNN/data/rna3db/test-pkl"
train_list_p = "drfold_train.list"
with open(train_list_p) as f:
    train_list = f.readlines()
train_pdbs1 = set([d[:4] for d in train_list])

rna3db_train = os.listdir(rna3db_train_dir)
train_pdbs2 = set([d[:4] for d in rna3db_train])

train_difference = train_pdbs1.difference(train_pdbs2)
print(len(train_difference), len(train_pdbs1), len(train_pdbs2))
print(len(train_difference)/len(train_pdbs1), len(train_difference)/len(train_pdbs2))
```

```python
test_files = "/home/mjustyna/data/rna3db-mmcifs/test-300"
test_pdbs = os.listdir(test_files)
test_pdbs = set([d[:4] for d in test_pdbs])
test_difference = test_pdbs.difference(train_pdbs1)
print(len(test_difference), len(test_pdbs), len(train_pdbs1))
print(len(test_difference)/len(test_pdbs))
```

```python
import shutil
test_pickles = "/home/mjustyna/RNA-GNN/data/rna3db/test-pkl"
out_test_path = "/home/mjustyna/RNA-GNN/data/rna3db-300/test-pkl"
test_files = os.listdir(test_pickles)

for file in test_files:
    pdb_name = file[:4]
    if pdb_name in test_difference:
        shutil.copy(os.path.join(test_pickles, file), os.path.join(out_test_path, file))
```
