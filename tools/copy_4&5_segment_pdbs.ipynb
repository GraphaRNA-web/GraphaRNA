```python
import os

seg_type = 5
desc_path = "/home/mjustyna/data/desc-pdbs/"

with open("all-output.txt", 'r') as f:
    lines = f.readlines()

lines = [l.split() for l in lines]
lines = [l for l in lines if len(l)>=7 and len(l[1])>=12]
pairs = dict([(l[1], int(l[2])) for l in lines])
seqs = dict([(l[1], l[-pairs[l[1]]:]) for l in lines])
pairs
```

```python
all_given_type = [f for f, v in pairs.items() if v == seg_type]
print(len(all_given_type))
all_given_type = [f for f in all_given_type if os.path.exists(desc_path + f +'.pdb')]
print(len(all_given_type))
```

```python
import random
random.seed(42)
random.shuffle(all_given_type)
```

```python
import shutil
out_dir = f'/home/mjustyna/data/eval_examples/{seg_type}_segment'
os.makedirs(out_dir, exist_ok=True)

for f in all_given_type[:32]:
    f += '.pdb'
    shutil.copy(os.path.join(desc_path, f), os.path.join(out_dir, f))
```
